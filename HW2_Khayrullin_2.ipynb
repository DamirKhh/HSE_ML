```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

data = pd.DataFrame()
for i in range(10):
  info = pd.read_csv(f"/content/Students_info_{i}.csv")
  marks = pd.read_csv(f"/content/Students_marks_{i}.csv")
  merged = pd.merge(info, marks, on = 'index')
  data = pd.concat([data,merged])
data.head()
```

**Задание №12**

```python
import time
start_time3 = time.time()
ethnics = data['race/ethnicity'].unique()
reading_avg = {}
for group in ethnics:
  avg = data[data['race/ethnicity'] == group]['reading score'].mean()
  reading_avg[group] = avg
print(reading_avg)
print(time.time() - start_time3)
print(time.time() - start_time3 - time_ans1)
```

```python
import time
start_time4 = time.time()
educations = data['education'].unique()
educ_min = {}
for level in educations:
  min = data[data['education'] == level]['writing score'].min()
  educ_min[level] = min
print(educ_min)
print(time.time() - start_time4)
print(time.time() - start_time4 - time_ans2)
```

**Задание №14**

```python
passmark = 50
data['Math_PassStatus'] = data['math score'].apply(lambda x: 'P' if x >= passmark else 'F')
data['Reading_PassStatus'] = data['reading score'].apply(lambda x: 'P' if x >= passmark else 'F')
data['Writing_PassStatus'] = data['writing score'].apply(lambda x: 'P' if x >= passmark else 'F')
print(data['Writing_PassStatus'].value_counts())
print(data['Reading_PassStatus'].value_counts())
print(data['Writing_PassStatus'].value_counts())
```

**Задание №15**

```python
data['OverAll_PassStatus'] = data.apply(lambda x: 'F' if x['Math_PassStatus'] == 'F' or x['Reading_PassStatus'] == 'F' or x['Writing_PassStatus'] == 'F' else 'P', axis = 1 )
print(data['OverAll_PassStatus'].value_counts())
```

**Задание №16**

```python
def GetGrade(average_mark):
  if average_mark > 90:
    return 'A'
  elif average_mark >= 80:
    return 'B'
  elif average_mark >= 70:
    return 'C'
  elif average_mark >= 60:
    return 'D'
  elif average_mark >= 50:
    return 'E'
  else:
    return 'F'
data['average_mark'] = (data['math score'] + data['reading score'] + data['writing score']) / 3
data['Grade'] = data['average_mark'].apply(GetGrade)
data = data.drop('average_mark', axis = 1)
data['Grade'].value_counts()
```
